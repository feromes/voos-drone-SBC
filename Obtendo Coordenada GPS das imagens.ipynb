```python
df_videos.CreateDate = pd.to_datetime(df_videos.CreateDate, format='%Y:%m:%d %H:%M:%S')
```

```python
def duracao(duration):
    if duration[-1] == 's':
        return pd.Timedelta(float(duration.replace(' s', '')), unit='s')
    else:
#         return np.sum(np.abs(duration.split(':')) * np.asarray((3600, 60, 1)))
        return pd.Timedelta(np.sum(np.fromstring(duration, dtype=int, sep=':') * np.asarray((3600, 60, 1))), unit='s')
```

```python
df_videos.Duration = df_videos.Duration.apply(lambda x: duracao(x))
```

```python
df_videos['final'] = df_videos.CreateDate + df_videos.Duration
```

```python
# df_videos
```

```python
# função de conversão de GMS (Grau, Minuto, Segundo) para ângulo decimal
def gms_para_decimal_SW(coord):
    coord = coord.replace('deg ', '').replace('\'', '').replace('" ', '').replace('S', '').replace('W', '')
    divisor = np.asarray((1, 60, 3600))
    gms = np.fromstring(coord, dtype=float, sep=' ')
    return np.sum(np.abs(gms / divisor)) * -1.0
```

```python
df_videos['lat'] = df_videos[df_videos.GPSLatitude.notna()].GPSLatitude.apply(lambda x: gms_para_decimal_SW(x))
```

```python
df_videos['long'] = df_videos[df_videos.GPSLongitude.notna()].GPSLongitude.apply(lambda x: gms_para_decimal_SW(x))
```

```python
gdf_videos = gpd.GeoDataFrame(df_videos, geometry=gpd.points_from_xy(df_videos.long, df_videos.lat))
```

```python
all_files = glob.glob('./log-voos/DJIFlightRecord*.csv')
print(all_files)
```

```python
li = []

# Carrega dados de cada um dos arquivos
print(f"Carregando os arquivos de log ...")
for filename in all_files:
    print(f" .... {filename}")
    name = filename.replace('./log-voos/', '').replace('.csv', '')
    df = pd.read_csv(filename, index_col=None, header=0, encoding="UTF-8", delimiter=',')
    df['Inicio'] = pd.to_datetime(name, format="DJIFlightRecord_%Y-%m-%d_[%H-%M-%S]")
    df['DateTime'] = df['Inicio'] + pd.to_timedelta(df['Time(seconds)'], unit='s')
    li.append(df)

# junta todos os anos em um único DataFrame
df = pd.concat(li, axis=0, ignore_index=True)
print(f"{len(df)} de registros carregados!")
```

```python
# Gerando base geográfica com Lat/Long
print('Gerando base geográfica com Lat/Long')
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))

# Definindo e convertendo SRC
print('Definindo e convertendo SRC')
gdf.crs = 'epsg:4989'
gdf = gdf.to_crs('epsg:31983')
```
